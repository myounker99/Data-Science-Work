The Birthday paradox is one of the most fasinating paradoxes out there and it can easily come up in conversation (alright well maybe not easily, but you can always work it in somehow)

https://en.wikipedia.org/wiki/Birthday_problem

Using the closed form solution found on wikipedia, replicate the plot using python code

```python
import matplotlib.pyplot as plt
import numpy as np

#x is num of people
#y is prob of birthday match

b_ls = list()

for i in range(0, 366):
    b_ls.append(1 - ((np.math.factorial(365))/(365**(i)*np.math.factorial(365-i))))
    
        
plt.plot(range(0,366), b_ls)
plt.xlabel("Number of people")
plt.ylabel("Probability of a match")
plt.title("Probability of Birthday Match (Derived Deterministically)")
plt.show()
```

```python
#though i don't use 0 or 1, making the list 365 instead of 363 makes indexing it easier
matches_per_n = [0] * 366

bd_ls = list()

#ignore 0 and 1, cannot match b-days with no one or one person
for n in range(2,366):
    #for 1000 people
    for i in range(0,1000):
        #generate b-days
        for j in range(0,n):
            bd_ls.append(np.random.randint(1,366))
        st = set(bd_ls)

        #if the length of the set is equal to the length of the list,
        #we know that there is no repeat in b-days
        if(len(st) != len(bd_ls)):
                matches_per_n[n] += 1
        bd_ls.clear()            
            
plt.plot(range(0,366), matches_per_n)
plt.xlabel("Number of people")
plt.ylabel("Probability of a match")
plt.title("Probability of Birthday Match (Derived Stochastically)")
plt.show()
```
